```python
dictionary = {}
for label, sector in df[['Label', 'Sector']].values:
    if label not in dictionary:
        if sector is not np.nan:
            dictionary[label] = sector
    
```

```python
# # set attributes
# for idx, node in enumerate(graph.nodes):
#     graph[idx]['sector'] = dictionary.get(node, np.nan)
nx.set_node_attributes(graph, dictionary, 'sector')
```
